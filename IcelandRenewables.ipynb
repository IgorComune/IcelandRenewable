## Data

```python
# showing all columns
pd.set_option('display.max_columns', None)
```

```python
#importing the dataset
all_data = pd.read_excel('owid-energy-data.xlsx')
#bringing all data from Iceland
iceland_data = all_data.query('country == "Iceland"')
```

```python
iceland_data.head()
```

```python
#drop NAN values
iceland_data.dropna(axis=1, inplace=True)
```

## reseting e setting index

```python
iceland_data.drop(labels=['iso_code','country'], axis=1, inplace=True)
```

```python
iceland_data.reset_index(inplace=True)
```

```python
iceland_data.set_index('year')
iceland_data.drop(labels='index', axis=1, inplace=True)
```

```python
iceland_data.set_index('year')
```
